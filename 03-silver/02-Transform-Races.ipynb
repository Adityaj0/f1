```python
bronze_table = f"{catalog_name}.{bronze_schema}.races"
silver_table = f"{catalog_name}.{silver_schema}.races"
```

```python
from pyspark.sql import functions as F
```

```python
races_df = spark.table(bronze_table)
```

```python
races_selected_df = races_df.select(F.col("season"), F.col("round"), F.col("raceName"), F.col("date"), F.col("circuitId"), F.col("ingested_at"), F.col("source_file"))
```

```python
races_renamed_df = (
    races_selected_df
    .withColumnsRenamed({
        "circuitId": "circuit_id",
        "raceName": "race_name",
        "date": "race_date"
    })
)
```

```python
races_distinct_df = races_renamed_df.dropDuplicates(["season", "round"])
```

```python
races_final_df = (
    races_distinct_df
    .withColumn("race_name", F.initcap(F.col("race_name")))
)
```

```python
(races_final_df
 .write
 .format("delta")
 .mode("overwrite")
 .saveAsTable(silver_table)
)
```

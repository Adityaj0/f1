```python
source_file = f"{landing_folder_path}/races.csv"
table_name = f"{catalog_name}.{bronze_schema}.races"
```

```python
from pyspark.sql.types import StructType, IntegerType, DateType, StringType, StructField

race_schema = StructType(fields=[
    StructField("season", IntegerType(), False),
    StructField("round", StringType(), True),
    StructField("url", StringType(), True),
    StructField("raceName", StringType(), True),
    StructField("date", DateType(), True),
    StructField("circuitId", StringType(), True)
])
```

```python
races_df = (
    spark.read
    .format("csv")
    .option("header", True)
    .option('mode', 'FAILFAST')
    #.option('inferSchema', True)
    .schema(race_schema)
    .load(source_file)
)
```

```python
display(races_df)
```

```python

races_final_df = add_ingestion_meta(races_df)
```

```python
display(races_final_df)
```

```python
(
    races_final_df
    .write
    .mode('overwrite')
    .format('delta')
    .saveAsTable(table_name)
)
```
